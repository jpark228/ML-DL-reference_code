# 리스트 (List)

```python
# mutable (불변) - element 변경 가능
x = list()
y = []

print(x)
print(y)
```

```python
x = [1, 2, 3, 4]
y = ["hello", "world"]
z = ["hello", 1,2,3]

print(x)
print(y)
print(z)
```

```python
print(x + y)
```

```python
x = [1, 2, 3, 4]

print(x[0])
print(x[3])
```

```python
x = [1, 2, 3, 4]

x[3] = 10
print(x)
```

```python
x = [1, 2, 3, 4]

num_elements = len(x)
print(num_elements)
```

```python
x = [4, 2, 3, 1]

y = sorted(x)
print(y)
```

```python
x = [4, 2, 3, 1]

z = sum(x)
print(z)
```

```python
x = [4, 2, 3, 1]

for n in x:
  print(n)
```

```python
y = ["hello", "there"]
for c in y:
  print(c)
```

```python
x = [4, 2, 3, 1]
print(x.index(3))
```

```python
y = ["hello", "there"]
print(y.index("hello"))
```

```python
y = ["hello", "there"]
print("hello" in y)
```

```python
y = ["hello", "there"]
print("bye" in y)
```

```python
y = ["hello", "there"]

if "hello" in y:
  print("hello 가 있어요")
```

```python
x = [4, 2, 3, 1]

if "hello" in x:
  print("hello 가 있어요")
```

# 튜플 (tuple)

```python
# immutable (불변) - element 변경 불가능
x = tuple()
y = ()

print(x)
print(y)
```

```python
x = (1, 2, 3)
y = ('a', 'b', 'c')
z = (1,"hello","there")

print(x)
print(y)
print(z)
```

```python
x = (1, 2, 3)
y = ('a', 'b', 'c')
z = (1,"hello","there")

print(x + y)
print('a' in y)
print(z.index(1))
```

# 딕셔너리 (Dictionary)

```python
 x = dict()
 y = {}

 print(x)
 print(y)
```

```python
x = {
  "name": "워니",
  "age": 20,
}

print(x)
```

```python
print(x["name"])
print(x["age"])
```

```python
x = {
  0: "Wonie",
  1: "Hello",
  "age": 20,
}

print(x[0])
print(x[1])
print(x["age"])
```

```python
print("age" in x)
```

```python
print("name" in x)
```

```python
x = {
  0: "Wonie",
  1: "Hello",
  "age": 20,
}

print(x.keys())
```

```python
print(x.values())
```

```python
for key in x:
  print("key: " + str(key))
  print("value: " + str(x[key]))
```

```python
x = {
  0: "Wonie",
  1: "Hello",
  "age": 20,
}

x[0] = "워니"
print(x)
```

```python
x = {
  0: "Wonie",
  1: "Hello",
  "age": 20,
}

x["school"] = "한빛"
print(x)
```
