# 조건문 (control - If Else)

```python
if 2 > 1:
  print("Hello")
```

```python
if 1 > 2:
  print("Hello")
```

```python
if not 1 > 2:
  print("Hello")
```

```python
if 1 > 0 and 2 > 1:
    print("Hello")
```

```python
if 0 > 0 and 2 > 1:
    print("Hello")
```

```python
if 0 > 0 or 2 > 1:
    print("Hello")
```

```python
x = 3

if x > 2:
    print("Hello")
```

```python
x = 3

if x > 5:
    print("Hello")
else:
    print("Hi")
```

```python
x = 3

if x > 5:
    print("Hello")
elif x == 3:
    print("Bye")
else:
    print("Hi")
```
