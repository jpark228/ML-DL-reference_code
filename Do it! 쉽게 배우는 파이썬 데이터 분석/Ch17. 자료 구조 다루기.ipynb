## 17-5 딕셔너리

#### 만들기

```python
# 딕셔너리 만들기
x = {'name' : 'kim'}
x
```

```python
type(x)
```

```python
x = {'id' : 1, 'name' : ['kim', 'park'], 'nclass' : (1, 2, 3)}
x
```

#### 추출하기

```python
# id 키의 값 추출
x['id']  
```

```python
# name 키의 값 추출
x['name'] 
```

```python
# name 키의 0번째 값 추출
x['name'][0]  
```

```python
# nclass 키의 1번째 값 추출
x['nclass'][1]  
```

#### 수정하기

**추가하기**

```python
# 'score': [5, 6, 7] 쌍 추가
x['score'] = [5, 6, 7]  
x
```

**삭제하기**

```python
# id 키:값 쌍 삭제
del x['id']  
x
```

```python
# name 키의 1번째 요소 삭제
del x['name'][1]  
x
```

**값 변경하기**

```python
# score 키의 값 변경
x['score'] = [7, 8, 9]  
x
```

```python
# score 키의 2번째 값 변경
x['score'][2] = 99  
x
```

**키 변경하기**

```python
# (1) 기존 키를 새 키에 할당
x['math'] = x['score']
x
```

```python
# (2) 기존 키 삭제
del x['score'] 
x
```

## 17-6 시리즈

#### 만들기

```python
# 시리즈 만들기
import pandas as pd
x_num = pd.Series([3, 4, 5])
x_num
```

```python
type(x_num)
```

```python
# 인덱스 정하기
x_label = pd.Series([3, 4, 5], index = ['kim', 'lee', 'park'])
x_label
```

#### 추출하기

```python
# 0번째 값 추출
x_num[0]  
```

```python
# 1번째 값 추출
x_num[1]  
```

```python
# 2번째 값 추출
x_label[2]      
```

```python
# 인덱스 kim의 값 추출
x_label['kim']  
```

```python
x_label.loc['kim']
```

```python
x_num.iloc[0]
```

#### 수정하기

**추가하기**

```python
# 3번째 인덱스에 9 할당
x_num[3] = 9  
x_num
```

```python
# 인덱스 choi에 6 할당
x_label['choi'] = 6  
x_label
```

**삭제하기**

```python
# 2번째 인덱스 삭제
del x_num[2]         
x_num
```

```python
# 인덱스 park 삭제
del x_label['park']  
x_label
```

**변경하기**

```python
# 3번째 인덱스의 값을 4로 변경
x_num[3] = 4        
x_num
```

```python
# 인덱스 kim의 값을 5로 변경
x_label['kim'] = 5  
x_label
```
