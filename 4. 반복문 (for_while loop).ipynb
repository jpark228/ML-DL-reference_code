# For loop

```python
for i in range(10):
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")
```

```python
for i in range(3):
  print(i)
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")
```

# While loop

```python
i = 0
while i < 3:
  print(i) 
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")
  i = i + 1 
```

# Break

```python
i = 0
while True:
  print(i) 
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")
  i = i + 1

  if i > 2:
    break
```

```python
for i in range(100):
  print(i) 
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")

  if i > 2:
    break
```

# Continue

```python
for i in range(3):
  print(i) 
  print("철수: 안녕 영희야 뭐해?")
  print("영희: 안녕 철수야, 그냥 있어.")
  
  continue

  print("워니: 안녕 철수화 영희야!")
```
